```python
import pandas as pd
import numpy as np
```

```python
# 初始化变量以存储所有csv文件的指标结果
r2_scores = []
maes = []
rmses = []

# 从csv中逐个读取并计算指标
for i in range(1, 73):  # csv文件名为data1.csv至data72.csv
    filename = r'./DLinear/results/road{}-EXPONENT-ForecastResults.csv'.format(i)
    df = pd.read_csv(filename)  # 读取CSV文件
    true_values = df['real']
    predicted_values = df['forecast']

    # 计算R平方值
    ss_res = np.sum((true_values - predicted_values) ** 2)
    ss_tot = np.sum((true_values - np.mean(true_values)) ** 2)
    r2 = 1 - (ss_res / ss_tot)
    r2_scores.append(r2)

    # 计算平均绝对误差MAE
    mae = np.mean(np.abs(true_values - predicted_values))
    maes.append(mae)

    # 计算均方根误差RMSE
    rmse = np.sqrt(np.mean((true_values - predicted_values) ** 2))
    rmses.append(rmse)

for i in range(72):
    print(f"在{filename}上的结果：")
    print(f"R方平方值：{r2_scores[i]}")
    print(f"平均绝对误差MAE：{maes[i]}")
    print(f"均方根误差RMSE：{rmses[i]}")

    
```

```python
def ListMeanStderror(my_list):
    
    # 将列表转换为NumPy数组
    my_array = np.array(my_list)

    # 计算均值
    mean_value = np.mean(my_array)

    # 计算标准差
    std_deviation = np.std(my_array)

    return ('均值为:', mean_value),('标准差为:', std_deviation)
```

```python
ListMeanStderror(r2_scores)
```

```python
ListMeanStderror(maes)
```

```python
ListMeanStderror(rmses)
```
